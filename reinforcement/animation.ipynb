```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
fig, ax = plt.subplots()
xdata, ydata = [], []
line, = ax.plot([], [], 'r-', animated=False)
def init():
   ax.set_xlim(0, 2 * np.pi) # 设置 x 轴范围
   ax.set_ylim(-1, 1) # 设置 y 轴范围
   return line,
```

```python
def update(frame):
   xdata.append(frame)
   ydata.append(np.sin(frame)) # 绘制正弦函数
   line.set_data(xdata, ydata)
   return line,
```

```python
frames = np.linspace(0, 2 * np.pi, 128) # 定义帧数
ani = FuncAnimation(fig, update, frames=frames, init_func=init, blit=True)
plt.show()
```

```python
ani.save('sine_wave.gif', writer='imagemagick', fps=30)
```
