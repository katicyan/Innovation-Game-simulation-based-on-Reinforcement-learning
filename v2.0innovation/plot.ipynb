```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import os

def plot_time_series(rh: pd.DataFrame) -> None:
    for i in range(10):
        plt.plot(rh[f"firm_{i}"], label=f"Agent {i} Revenue")
    plt.xlabel("Episode")
    plt.ylabel("Value")
    plt.title("Value History of Agents")
    plt.legend()
    plt.show()
```

```python
output_dir = os.path.join(os.getcwd(), "v2.0innovation", "experiment_data")
revenue_csv = os.path.join(output_dir, "revenue_history.csv")
action_csv = os.path.join(output_dir, "action_history.csv")

rh = pd.read_csv(revenue_csv)
ah = pd.read_csv(action_csv)
```

```python
# plot_time_series(rh)
plot_time_series(ah[0:50])
plot_time_series(ah[50:100])
plot_time_series(ah[100:150])
plot_time_series(ah[150:200])
```
