```python
import argparse
# Example: parse --name and --repeat from command line
parser = argparse.ArgumentParser(description="Simple argparse example")
parser.add_argument("--name", type=str, default="World", help="Name to greet")
parser.add_argument("--repeat", type=int, default=1, help="How many times to print")
def parse_list(value):
    return [int(x.strip()) for x in value.strip("[]").split(",") if x.strip()]
parser.add_argument(
    "--work-hours",
    type=parse_list,
    default=[9, 17],
    help="Working hours as a list of numbers"
)
# In Jupyter, use parse_args([]) or provide a custom list of args
args = parser.parse_args(["--name", "Alice", "--repeat", "3", "--work-hours", "[8,16,10]"])
# For normal Python scripts, use:
# args = parser.parse_args()

for _ in range(args.repeat):
    print(f"Hello, {args.name}!")
```

```python
print(f"Work hours: {args.work_hours}")
```

```python
import os
import pickle

# Save the existing `args` object to a pickle file
with open("args.pkl", "wb") as f:
    pickle.dump(args, f)

# Load it back
with open("args.pkl", "rb") as f:
    loaded_args = pickle.load(f)

print("Original:", args)
print("Loaded:  ", loaded_args)
print("Same values:", vars(args) == vars(loaded_args))
```

```python
class MathUtils:
    def __init__(self):
        self.infty = float('inf')

    def division(self, a, b):
        try:
            return a / b
        except ZeroDivisionError:
            return self.infty

    @staticmethod
    def add(a, b):
        return a + b


print(MathUtils.add(2, 3))
```

```python
obj = MathUtils()
print(obj.division(10, 0))  # Should print inf
```
